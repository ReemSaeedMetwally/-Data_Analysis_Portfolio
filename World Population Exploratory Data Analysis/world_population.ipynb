```python
#grouping by continent and sorting all the mean values in a descending order according to 2022 Population
df2 = df.groupby('Continent').mean(numeric_only = True).sort_values(by ="2022 Population",ascending = False)
df2
```

```python
#choosing the columns of intresent from the DataFrame df2
#note that we need to have the years in an ascending order to have a better visulaization

df3 = df.groupby('Continent')[["1970 Population", "1980 Population", "1990 Population", "2000 Population","2010 Population", 
                               "2015 Population", "2020 Population", "2022 Population"]].mean().sort_values(by="2022 Population",ascending=False)
df3
```

```python
#Another way that we could use the index of the columns of intresent from the DataFrame df2
#note that we used the expression [::-1] to reverse the range in order to have the years in an ascending order to have a better visulaization

df4 = df.groupby('Continent')[df.columns[5:13][::-1]].mean().sort_values(by="2022 Population",ascending=False)
df4
```

```python
#transposing the DataFrame df3 (swapping its rows and columns) for a better visualization
df5 = df3.transpose()
df5
```

```python
#visualizing the population of each continent by years
df5.plot()
```
